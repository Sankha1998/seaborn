```
# Problem 2: Plot a pairs plot on iris data set showing different species. Write your conclusions as well.
iris = sns.load_dataset('iris')
iris.head()
```

```
sns.pairplot(iris,hue='species')
```


petal_length vs petal_width helps to draw a conclusion on the identification of of the flowers.

```
# Problem 4: Plot a pie chart of survived based on sex.
titanic=sns.load_dataset('titanic')
titanic.head()
```
